```python
# The following code imports tools we need.
# It creates a dataframe (similar to an excel table)
# We then clean up the column names and assign the stats
# that we are interested in

import pandas as pd
df = pd.read_csv('projections/FantasyPros_2015.csv')
cols = df.columns.values
for i in range(len(cols)):
    cols[i] = cols[i].strip()
df.columns = cols
array_point_stats=['runs', 'rbi', 'hrs', 'sb', 'ave']
array_correlated_stats=['runs', 'rbi', 'hrs']
grease = pd.read_csv('grease.txt')

unavailable_players = []
```

```python
# This function compares a player to a dataframe of players.
# Each cell where the player beats a player in the dataframe
# is marked as true. Each cell where a player loses to a player in the
# dataframe is marked as false. We sum all of these cells and get a score.
# 
# The score is how many points they earn against the entire field of
# players.

def compete(player, data_frame, array_stats):
    t = player[array_stats] >= data_frame[array_stats]
    score = t[array_stats].sum().sum()
    return score
```

## Ranking
The following table is based off 2015 projections from [FantasyPros.com](http://www.fantasypros.com/mlb/projections/hitters.php)

The score on the rightmost side of the table is calculated by having each player compete against each other player
for each of the five hitting statistics. Each time the player beats someone else's stat, a point was added
to their score.

The table is sorted in order of score.

```python
df['score'] = df.apply(compete, data_frame = df, array_stats = array_point_stats, axis=1)
df = df.sort(columns='score', ascending=False)
df.head(10)
```

## Determining keepers
Here I use the index to keep track of the ranking of people according to the ADP, then create a new *erank* which is my ranking of them.
Then I calculate the difference between my rank and ADP and list them in order of greatest difference.

```python
df['adp'] = df.index.values + 1
df.index = range(0, len(df.index))
df['point_stats_rank'] = df.index.values + 1
df['diff_adp_point_stats'] = df['adp'] - df['point_stats_rank']
df.head()
```
