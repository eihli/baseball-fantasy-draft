# The Backstop Boys League Draft Analysis

### First we import pandas and read in the Batting.csv file from the [lahman database](http://www.seanlahman.com/baseball-archive/statistics/)

```python
import pandas as pd
```

```python
df = pd.read_csv('lahman/Batting.csv')
```

```python
df.head()
```

### Now we will select only data from years 2012 onwards.
### Then we will select only the relevant columns: R, HR, SB, AVG, RBI

```python
df = df[df['yearID'] >= 2012]
df.head()
```

```python
df['AVG'] = df['H'] / df['AB']
df = df[['playerID', 'yearID', 'R', 'HR', 'SB', 'AVG', 'RBI']]
df.head()
```

```python
(df['R'] - df['R'].mean()).head()
```

### You can get the mean of multiple columns at once.

```python
(df[['R', 'HR']].mean()).head()
```

### Pandas automatically ignores NaN values when you call something like mean()

```python
(df[['R', 'HR', 'AVG']].mean()).head()
```

```python
# Normalize the data. Subtract the mean of each column from each cell in that column, then divide that number
# by the standard deviation of the column.
# This might not be exactly what we want since the stats are heavily '0' weighted. I don't think this is what you
# would call a 'normal distribution'?
# It might still work anyway. It's good for now at the very least.

((df.iloc[:,2:] - df.iloc[:,2:].mean()) / df.iloc[:,2:].std()).head()
```

```python
df_norm = df
df_norm.iloc[:,2:] = (df.iloc[:,2:] - df.iloc[:,2:].mean()) / df.iloc[:,2:].std()
```

```python
df_norm.head()
```
