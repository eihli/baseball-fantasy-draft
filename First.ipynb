# The Backstop Boys League Draft Analysis

### First we import pandas and read in the Batting.csv file from the [lahman database](http://www.seanlahman.com/baseball-archive/statistics/)

```python
import pandas as pd
```

```python
df = pd.read_csv('lahman/Batting.csv')
df_master = pd.read_csv('lahman/Master.csv')
df_master = df_master[['playerID', 'nameFirst', 'nameLast']]
```

```python
df.head()
```

### Now we will select only data from years 2012 onwards.
### Then we will select only the relevant columns: R, HR, SB, AVG, RBI

```python
df = df[df['yearID'] >= 2012]
df['AVG'] = df['H'] / df['AB']
df = df[['playerID', 'yearID', 'R', 'HR', 'SB', 'AVG', 'RBI']]
df.head()
```

### Pandas automatically ignores NaN values when you call something like mean()

```python
(df[['R', 'HR', 'AVG']].mean()).head()
```

```python
# Normalize the data.
```

```python
df_stats = df[['R', 'HR', 'SB', 'AVG', 'RBI']]
df_stats_norm = df_stats / df_stats.sum()
df_stats_norm.head()
df_norm = pd.concat([df[['playerID', 'yearID']], df_stats_norm], 1)
df_norm.head()
```

```python
df_norm['sum'] = df_norm.iloc[:,2:].sum(1)
df_norm.head()
```

```python
player_averages = df_norm.groupby('playerID')['sum'].mean() 
player_averages.head()
player_averages.sort(ascending=False)
df_pa = pd.DataFrame([player_averages.keys(), player_averages.values]).transpose()
df_pa.columns = ['playerID', 'sum']
df_pa = df_pa.merge(df_master, on='playerID', how='inner')
df_pa.head(20)
```

```python
df_norm = df_norm.sort(['sum'], ascending=False)
df_norm.head()
```

We want to use the quantile function to select all rows where the player is in the top x percent.

```python
df_norm = df_norm[(df_norm.iloc[:,2] > df_norm.iloc[:,2].quantile(.95)) &
        (df_norm.iloc[:,6] > df_norm.iloc[:,6].quantile(.95)) &
        (df_norm.iloc[:,4] > df_norm.iloc[:,4].quantile(.95))]
```

```python
df_norm.shape
```

```python
pd.merge(df_norm, df_master, on='playerID', how='inner').head()
df_norm.sort('sum', ascending=False).head()
```

```python
groups = df_norm.groupby('playerID')
df_player_averages = groups['sum'].mean()
df_player_averages.shape
```

```python
df_norm['sum3ya'] = df_player_averages
df_norm.sort('sum3ya', ascending=False)
df_norm.head()
```
