```python
import pandas as pd
df = pd.read_csv('projections/FantasyPros_2015_P.csv')
cols = df.columns.values
for i in range(len(cols)):
    cols[i] = cols[i].strip()
df.columns = cols
```

```python
def compete(player, bowling, golf):
    bowling_df = player[bowling] >= df[bowling]
    golf_df = player[golf] <= df[golf]
    bowling_score = bowling_df.sum().sum()
    golf_score = golf_df.sum().sum()
    score = bowling_score + golf_score
    return score
df.head()
```

```python
df.head().apply(compete, bowling=['w', 'sv', 'k'], golf=['whip', 'era'], axis=1)
```

This table is based off 2015 projections from [FantasyPros.com](http://www.fantasypros.com/mlb/projections/hitters.php)

The score on the rightmost side of the table is calculated by having each player compete against each other player
for each of the five hitting statistics. Each time the player beats someone else's stat, a point was added
to their score.

The table is sorted in order of score.

```python
df['score'] = df.apply(compete, bowling=['w', 'sv', 'k'], golf=['whip', 'era'], axis=1)
df = df.sort(columns='score', ascending=False)
df.head(20)[['Player Name'
             ,'Team'
             ,'Position'
             ,'whip'
             ,'k'
             ,'sv'
             ,'era'
             ,'w']]
```

## Determining keepers
Here I use the index to keep track of the ranking of people according to the ADP, then create a new erank which is my ranking of them.
Then I calculate the difference between my rank and ADP and list them in order of greatest difference.

```python
df['adp'] = df.index.values + 1
df.index = range(0, len(df.index))
df['erank'] = df.index.values + 1
df.head()
```

```python
df['diff_adp_erank'] = df['adp'] - df['erank']
```
